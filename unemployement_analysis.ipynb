```python
a= a.dropna()
```

```python
a.isnull().sum()
```

```python
x=a['Region']
y=a[' Estimated Employed'].values
plt.scatter(x, y, color = 'g',s = 100)
plt.xticks(rotation = 25)
plt.xlabel('Region')
plt.ylabel('Estimate Employed')
plt.title('Employment rate in given region', fontsize = 10)

plt.show()
```

```python
plt.plot(x, y, color = 'g', linestyle = 'dashed',
         marker = 'o',label = "Employment")

plt.xticks(rotation = 25)
plt.xlabel('Region')
plt.ylabel('Estimate employment')
plt.title('Employment rate in given region', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.hist(a[' Estimated Unemployment Rate (%)'], bins=25, alpha=0.45, color='blue')


plt.title("histogram ")

plt.legend(['Estimated Unemployement Rate %'])

plt.show()
```
